```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

df = df[["Country Name", "2022"]]

df = df.dropna()

df["2022"] = df["2022"].astype(int)

top10 = df.sort_values(by="2022", ascending=False).head(10)

plt.figure(figsize=(12,6))
plt.bar(top10["Country Name"], top10["2022"],
        color="steelblue", edgecolor="black")

plt.title("Top 10 Most Populous Countries (2022)", fontsize=16, color='blue')
plt.xlabel("Country", fontsize=12, color='brown')
plt.ylabel("Population", fontsize=12, color='brown')
plt.xticks(rotation=45, ha='right')

for i, value in enumerate(top10["2022"]):
    plt.text(i, value, f'{value:,}', ha='center', va='bottom', fontsize=9)

plt.tight_layout()
plt.show()
```
